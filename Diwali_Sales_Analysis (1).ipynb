### Q1] Based on the Diwali Sales Data, which genders are the majority of buyers, and which gender have higher purchasing power?

```python
df.columns
```

```python
re=sns.countplot(x='Gender',data=df)
for bars in re.containers: #Iterate over the containers (groups of bars) in the countplot 're'.
  re.bar_label(bars)# Add labels to each bar in the container, displaying the count.

# Created a countplot using Seaborn to visualize the distribution of 'Gender' in the DataFrame 'df'.
# sns.countplot is a function from the Seaborn library that is used to create a bar plot that shows the counts of observations in each categorical bin using bars
#'x='Gender'' specifies that the 'Gender' column will be plotted on the x-axis.
#'data=df' indicates that the data for the plot is coming from the DataFrame 'df'.
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
# Group the DataFrame 'df' by the 'Gender' column.
# The 'as_index=False' argument ensures that 'Gender' is not set as the index of the resulting grouped DataFrame.
# Calculate the sum of the 'Amount' column for each gender group.
# Sort the resulting DataFrame by the 'Amount' column in descending order
# (highest amount at the top)
```

```python
sales_re=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_re)
# x='Gender'=Specify 'Gender' to be plotted on the x-axis.
#y='Amount' =Specify the column 'Amount' to be plotted on the y-axis.
# data=sales_re = Use the DataFrame 'sales_re' as the data source for the plot.
```

###AND IF YOU WANT TO ADD LABEL THO THE BARS PLOTED YOU CAN USE THIS CODE

```python
sales_re=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ffh=sns.barplot(x='Gender',y='Amount',data=sales_re) # we have assign this barlopt too ffh
for bars in ffh.containers:
  ffh.bar_label(bars)

# x='Gender'=Specify 'Gender' to be plotted on the x-axis.
#y='Amount' =Specify the column 'Amount' to be plotted on the y-axis.
# data=sales_re = Use the DataFrame 'sales_re' as the data source for the plot.
```

# CONCLUSION:-
### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
